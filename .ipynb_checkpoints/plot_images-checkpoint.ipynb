### Load a saved model

```python
from collections import OrderedDict

def load_to_model(model, saved_model_path):
    
    checkpoint = torch.load(saved_model_path)
    new_state_dict = OrderedDict()
    for k, v in checkpoint['state_dict'].items():
        name = k[7:] # remove `module.`
        new_state_dict[name] = v
    # load params
    model.load_state_dict(new_state_dict)
    return model

```

```python
def draw_sample_images(images, num_samples, filename=None):
    
    plt.figure(figsize=(8, 9))
    plt.suptitle('Image generated with xxx models', fontsize=16)
    plt.subplots_adjust(left=.05, right=1, bottom=.001, top=.96, wspace=.2,
                hspace=.2)
    
    if isinstance(images, list):
        num_models = len(images)
    else:
        num_models = 1
        images = [images]
    
    num_samples = images[0].size()[0]
    
    
    for i in range(0, num_samples):
        for j in range(1, num_models+1):
            plt.subplot(num_samples, num_models, i * num_models + j)
            img = images[j-1][i]
            plt.axis('off')
            plt.imshow(img.permute(1, 2, 0).detach().numpy())
        
    if filename is not None:
        prefix='./pics/'
        suffix='.png'
        plt.savefig(prefix+filename+suffix)

    plt.show()
```

```python
model = VAE(z_dim=128)
path = './models/vae_20_checkpoint.pth.tar'
model = load_to_model(model, path)
```

```python
pics = model.generate(3, None)
pics2 = model.generate(3, None)
```

```python
draw_sample_images([pics, pics2], 3)
```
